stroke:

Establecen el color utilizado para dibujar líneas y bordes alrededor de las formas. Este color se especifica en términos del color RGB o HSB dependiendo del colorMode actual () (el espacio de color predeterminado es RGB, con cada valor en el rango de 0 a 255). El rango alfa por defecto también es de 0 a 255.

strokeweight:

Establece el ancho del trazo utilizado para líneas, puntos y el borde alrededor de las formas. Todos los anchos se establecen en unidades de píxeles.

El objetivo de la función es que se pueda cambiar los colores al gusto de la persona.


beginShape and endShape:

El uso de las funciones beginShape () y endShape () permite crear formas más complejas. beginShape () comienza a grabar vértices para una forma y endShape () detiene la grabación. El valor del parámetro kind le dice qué tipos de formas crear a partir de los vértices proporcionados. Sin modo especificado, la forma puede ser cualquier polígono irregular.

vertx(x,y):

Todas las formas se construyen conectando una serie de vértices. 
vertex () se utiliza para especificar las coordenadas de vértice para puntos, líneas, triángulos, quads y polígonos. Se usa exclusivamente dentro de las funciones beginShape () y endShape ().

background():


La función background () establece el color utilizado para el fondo del lienzo p5.js. El fondo predeterminado es transparente.

fill():

Establece el color utilizado para rellenar formas. Por ejemplo, si ejecuta el relleno (204, 102, 0), todas las formas dibujadas después del comando de relleno se rellenarán con el color naranja. Este color se especifica en términos de color RGB o HSB dependiendo del colorMode () actual.

nofill():

Inhabilita la geometría de relleno. Si se llama a noStroke () y noFill (), no se dibujará nada en la pantalla.

método crear_vertices:

Dibuja una elipse (ovalada) en la pantalla. Una elipse con igual ancho y alto es un círculo.

El número de veces que se ejecuta draw () en cada segundo puede controlarse con la función frameRate ().
